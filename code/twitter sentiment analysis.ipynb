```python
import pandas as pd
```

```python
df = pd.read_csv('G:/Learning/My jupyter csv files/tweet obama.csv')
df.head()
```

```python
from nltk.tokenize import RegexpTokenizer

pattern = RegexpTokenizer('\w+')

# we use \w+ for match unicode characters

df['after_tokenize'] = df['text'].apply(pattern.tokenize)
df.head()
```

```python
# stopwords are like pronouns, conjunctions, preprosition etc. they don't contain valuable information
# so we have to remove them

import nltk
from nltk.corpus import stopwords
stop_words = nltk.corpus.stopwords.words('english')

# custome stopwords

sw = ['https', 'http']
stop_words.extend(sw)
```

```python
# using lamdba remove stopwords

df['after_tokenize'] = df['after_tokenize'].apply(lambda x: [item for item in x if item not in stop_words])
df.head()
```

```python
# Remove Infrequent words
# infrequent or rare words are noisy. They create a kind of noise between common words and rare words.

df['rmv_infrequent_words'] = df['after_tokenize'].apply(lambda x: ' '.join([word for word in x if len(word)>2]))
df.head()
```

```python
# join all words

all_words = ' '.join([word for word in df['rmv_infrequent_words']])
all_words
```

```python
word_tokenize = nltk.tokenize.word_tokenize(all_words)
```

```python
from nltk.probability import FreqDist

freq_distribution = FreqDist(word_tokenize)
freq_distribution
```

```python
# it's a dictionary form.
# now we can use frequencey distribution and remove some words which are less necessary
```

```python
df['drop_words'] = df['after_tokenize'].apply(lambda x: ' '.join([word for word in x if freq_distribution[word]>3]))
df.head()
```

```python
df[['text', 'after_tokenize', 'rmv_infrequent_words', 'drop_words']].head()
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemma = WordNetLemmatizer()

df['after_lemmatization'] = df['drop_words'].apply(lemma.lemmatize)
df.head()
```

```python
df['is_equal'] = (df['drop_words']==df['after_lemmatization'])
```

```python
df.is_equal.value_counts()
```

```python
lemma_words = ' '.join([word for word in df['after_lemmatization']])
lemma_words
```

```python
words = nltk.word_tokenize(lemma_words)
fd = FreqDist(words)

most_common = fd.most_common(5)
most_common
```
